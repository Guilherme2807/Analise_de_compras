#### Importando bibliotecas

```python
# a biblioteca os irá ser responsável por ler a pasta com o arquivo, enquanto a parte de análise ficará com o pandas
import os
import pandas as pd
```

```python
# Sem o "r" antes do caminho, os caracteres de escape\,\n interpretado como uma quebra de linha, e \t interpretado como um caractere de tabulação. No entanto, quando você usa uma string de "raw" com o "r", os caracteres de escape são tratados literalmente.
#criando variável com o caminho da pasta
pasta = (r"C:/Users/guilh/OneDrive/Área de Trabalho/Agenda")
#listando as pastas
arquivos = os.listdir(pasta)
print(arquivos)
```

```python
# para eu ler a informação da planilha por meio do pandas, eu preciso criar uma variável que recebe um dataframe
# Construa o caminho completo do arquivo
caminho_arquivo = os.path.join(pasta, 'Empresa A.xlsx')

licitacao = pd.read_excel(caminho_arquivo)
display(licitacao)

#renomeando o nome total pela empresa
licitacao = licitacao.rename(columns={'Total': 'Empresa A'})
display(licitacao)

#como eu quero olhar apenas ao total de todas as empresas, o valor unitário perde sua utilidade, neste caso eu posso retirar o mesmo
licitacao = licitacao.drop(['Valor Unitário'], axis =1)
display(licitacao)

##Para incoporar as demais empresas ao final de minha tabela, eu preciso criar uma estrutura de repetição que permita isso
#porém dentro de minha pasta eu tenho arquivos que não são empresas, neste caso preciso pegar apenas até as empresas
empresas = arquivos [1:8]

for empresa in empresas:
    caminho_arquivo_2 = os.path.join(pasta, empresa)
    proposta = pd.read_excel(caminho_arquivo_2)
    licitacao[empresa[:-5]] = proposta['Total'].copy()
    
display(licitacao)
```

#### Verificando valores

```python
# tentando pegar o valor total de cada um dos fornecedores, para isso eu crio uma estrutura de repetição
for coluna in licitacao.columns[3:]:
    total = licitacao[coluna].sum()
    print('{} : R$ {:,.2f}'.format(coluna, total))
```

#### Análise e possíveis ações

### Resultado ótimo

```python
#o resultdo ótimo consiste em pegar o valor menor de cada empresa por item e montar um norte de referência
# para fazer isso eu crio uma coluna chamada mínima, retiro as três primeiras colunas, e realizo uma análise mínimo das colunas de empresas
licitacao ['Mínimo'] = licitacao.drop(['Item', 'Descrição do Item', 'Quantidade'], axis = 1).min(axis=1)
display(licitacao)


minimo =licitacao['Mínimo'].sum()
print('R${:,.2f}'.format(minimo))
```

### Planilha de desvios 1º parte

```python
#essa planilha busca apresentar o quão distante é cada item das empresas para coluna de mínimo
for empresa in licitacao.columns[3:-1]:
    #criando uma planilha que contenha as seguintes colunas
    desvio = licitacao[['Item', 'Descrição do Item', 'Quantidade', empresa,'Mínimo']]
    #criando uma coluna de delta para a planilha criada
    desvio['Delta'] = desvio[empresa]-desvio['Mínimo']
    #criando um arquivo excel com esses dados
    desvio.to_excel('Desvio - '+empresa+'.xlsx', index = False)
```

### Planilha de Desvio 2º Parte

```python
#agora na segunda parte eu irei fazer a junção dos desvios em uma planilha
pasta_2 = (r"C:/Users/guilh/anaconda3/Projetos Meus/pasta oficial")
arquivo_2=os.listdir(pasta_2)
```

```python
desvio_arquivo = arquivo_2[4:12]
print(desvio_arquivo)
```

```python
caminho_arquivo_12 = os.path.join(pasta_2, 'Desvio - Empresa A.xlsx')
desvio_licitacao = pd.read_excel(caminho_arquivo_12)
display(desvio_licitacao)
```

```python
#como eu quero olhar apenas o desvio da empresa e o minimo, os demais itens eu posso retirar 
desvio_licitacao = desvio_licitacao.drop(['Descrição do Item','Quantidade','Empresa A'], axis =1)
display(desvio_licitacao)

##Para incoporar as demais empresas ao final de minha tabela, eu preciso criar uma estrutura de repetição que permita isso
#porém dentro de minha pasta eu tenho arquivos que não são empresas, neste caso preciso pegar apenas até as empresas
empresa_desvios = desvio_arquivo[1:12]
for empresa_desvio in empresa_desvios:
    caminho_arquivo_13 = os.path.join(pasta_2, empresa_desvio)
    proposta_desvio = pd.read_excel(caminho_arquivo_13)
    desvio_licitacao[empresa_desvio[8:-5]] = proposta_desvio['Delta'].copy()

#renomeando o nome Delta e Mínimo pela Empresa A e menor preço
desvio_licitacao = desvio_licitacao.rename(columns={'Mínimo': 'Menor Preço'})
desvio_licitacao = desvio_licitacao.rename(columns={'Delta': 'Empresa A'})
display(desvio_licitacao)
```

### Verificando a diferença entre as empresas e o valor mínimo

```python
# tentando pegar de desvio de cada um dos fornecedores, para isso eu crio uma estrutura de repetição
for coluna in desvio_licitacao.columns[3:]:
    total = desvio_licitacao[coluna].sum()
    print('{} : R$ {:,.2f}'.format(coluna, total))
```
